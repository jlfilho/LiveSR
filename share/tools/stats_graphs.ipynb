## 1.2 QoE

```python
# filter only QoE metrics
data_filter = ((scores['Metrics']==group1) | (scores['Metrics']==group2) | (scores['Metrics']==group3) | (scores['Metrics']==group4))
score2=scores[data_filter]
score2 = score2.reset_index(drop=True)
#Calc max mean by rule and group 
maxQoE = score2.groupby(['abr Rule','Metrics']).agg({'Average value':['mean']}).max()
print(maxQoE)

# Normalize the average value
score2['Average value'] = score2['Average value'].apply(lambda x: x/maxQoE).values
score2.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme(style="whitegrid")
#sns.set_palette("dark")

abr_filter = ['LoL+','BOLA','L2A-LL']

fig, bar = plt.subplots(figsize = (11,7))
ax = sns.barplot(x="Metrics", y="Average value",hue="abr Rule", data=score2,hue_order=abr_filter,
                 capsize=.05,ci=95,errwidth=0.7)
ax.set_xlabel("Experienced scenarios")
ax.set_ylabel("Normalized QoE")
plt.show()
ax.figure.savefig("./graphs/main_QoE"+".eps",bbox_inches="tight")
```

```python
import math 
import scipy.stats as st
import numpy as np

def conf_intervals(stats,ci=0.95):
    ci_hi = []
    ci_lo = []
    for i in stats.index:
        m, c, s = stats.loc[i]
        if(c >= 30):
            ci_=st.norm.interval(alpha=ci, loc=m, scale=s/np.sqrt(c))
        else:
            ci_=st.t.interval(alpha=0.95, df=c-1, loc=m, scale=s/np.sqrt(c))
        ci_hi.append(ci_[1])
        ci_lo.append(ci_[0])

    stats['ci_hi'] = ci_hi
    stats['ci_lo'] = ci_lo
    return stats

stats = score2.groupby(['abr Rule','Metrics']).agg(['mean', 'count', 'std'])
ci = 0.95
stats = conf_intervals(stats,ci=0.95)
print(stats)

fig, bar = plt.subplots(figsize = (11,7))
ax = sns.barplot(x="Metrics", y="Average value",hue="abr Rule", data=score2,hue_order=abr_filter,
                 capsize=.05,ci=ci*100,errwidth=0.7)
ax.set_xlabel("Experienced scenarios")
ax.set_ylabel("Normalized QoE")
plt.show()
```

```python
import math 
import scipy.stats as st
import numpy as np

def conf_intervals(stats,ci=0.95):
    ci_hi = []
    ci_lo = []
    for i in stats.index:
        m, c, s = stats.loc[i]
        if(c >= 30):
            ci_=st.norm.interval(alpha=ci, loc=m, scale=s/np.sqrt(c))
        else:
            ci_=st.t.interval(alpha=0.95, df=c-1, loc=m, scale=s/np.sqrt(c))
        ci_hi.append(ci_[1])
        ci_lo.append(ci_[0])

    stats['ci_hi'] = ci_hi
    stats['ci_lo'] = ci_lo
    return stats

stats = result[['DL_bitrate','tracer']].groupby(['tracer']).agg(['mean', 'count', 'std'])
ci = 0.95
stats = conf_intervals(stats,ci=0.95)
print(stats)

fig, bar = plt.subplots(figsize = (11,7))
ax = sns.barplot(x="tracer", y="DL_bitrate", data=result,
                 capsize=.05,ci=ci*100,errwidth=0.7)
ax.set_xlabel("traces")
ax.set_ylabel("BW(kbps)")
plt.show()
```
