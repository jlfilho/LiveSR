# Normal discrete

```python
%matplotlib notebook
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean=10
std=2
size=10000
r = norm.rvs(loc=mean,scale=std,size=size)
#r = np.random.normal(mean, std, size)
ax.hist(np.floor(r), density=True, bins = 15, histtype='stepfilled', alpha=0.8)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
%matplotlib notebook
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.arange(0, 20)
mean=10
std=2
size=10000
prob = norm.pdf(x, mean, std)
prob = prob / prob.sum() # normalize the probabilities so their sum is 1
nums = np.random.choice(x, size = size, p = prob)
#print(nums)
plt.hist(nums, bins = 15)
```

# Zipf

```python
%matplotlib notebook
from scipy.stats import zipf
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean=0
a=1.5
size=60*60*60
r = zipf.rvs(a,loc=mean,size=size)
ax.hist(r, density=True, bins = 50, histtype='stepfilled', alpha=0.8)
ax.legend(loc='best', frameon=False)
plt.show()
```

# Teste

```python
from scipy.stats import zipf
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1)
a = 3
mean, var, skew, kurt = zipf.stats(a, moments='mvsk')
mean, var, skew, kurt
```

```python
x = np.arange(zipf.ppf(0, a),
              zipf.ppf(0.99, a))
ax.plot(x, zipf.pmf(x, a), 'bo', ms=8, label='zipf pmf')
ax.vlines(x, 0, zipf.pmf(x, a), colors='b', lw=5, alpha=0.5)
x
```

```python
rv = zipf(a)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```
